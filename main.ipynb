```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Read CSV file
filepath = os.path.join(os.getcwd(), 'sample_data', 'employee_promotion.csv')
employee_promotions = pd.read_csv(filepath)
```

```python
# Drop region and employee ID columns
employee_promotions.drop(labels=None, axis=1, index=None, columns=["region", "employee_id" ], level=None, inplace=True, errors='raise')
# Drop rows with any NaN values
employee_promotions.dropna(how='any')
```

```python
# Employee promotions per department 

#Filtering by department 
employee_sales = employee_promotions[employee_promotions['department'] == 'Sales & Marketing']
#Calculating percentage of men in the department that were promoted
sale_promotion_m = len(employee_sales[(employee_sales['gender'] == 'm') & (employee_sales['is_promoted']==1)]) / len(employee_sales[employee_sales['gender'] == 'm'])
#Calculating percentage of women in the department that were promoted 
sale_promotion_f = len(employee_sales[(employee_sales['gender'] == 'f') & (employee_sales['is_promoted'])]) / len(employee_sales[employee_sales['gender'] == 'f'])

employee_operations = employee_promotions[employee_promotions['department'] == 'Operations']
operations_promotion_m = len(employee_operations[(employee_operations['gender'] == 'm') & (employee_operations['is_promoted'])]) / len(employee_operations[employee_operations['gender'] == 'm'])
operations_promotion_f = len(employee_operations[(employee_operations['gender'] == 'f') & (employee_operations['is_promoted'])]) / len(employee_operations[employee_operations['gender'] == 'f'])

employee_tech = employee_promotions[employee_promotions['department'] == 'Technology']
tech_promotion_m = len(employee_tech[(employee_tech['gender'] == 'm') & (employee_tech["is_promoted"])]) / len(employee_tech[employee_tech['gender'] == 'm'])
tech_promotion_f = len(employee_tech[(employee_tech['gender'] == 'f') & (employee_tech["is_promoted"])]) / len(employee_tech[employee_tech['gender'] == 'f'])

employee_analytics = employee_promotions[employee_promotions['department'] == 'Analytics']
analytics_promotion_m = len(employee_analytics[(employee_analytics['gender'] == 'm') & (employee_analytics['is_promoted'])]) / len(employee_analytics[employee_analytics['gender'] == 'm'])
analytics_promotion_f = len(employee_analytics[(employee_analytics['gender'] == 'f') & (employee_analytics['is_promoted'])]) / len(employee_analytics[employee_analytics['gender'] == 'f'])

employee_rd = employee_promotions[employee_promotions['department'] == 'R&D']
rd_promotion_m = len(employee_rd[(employee_rd['gender'] == 'm') & (employee_rd['is_promoted'])])/len(employee_rd[employee_rd['gender'] == 'm'])
rd_promotion_f = len(employee_rd[(employee_rd['gender'] == 'f') & (employee_rd['is_promoted'])])/len(employee_rd[employee_rd['gender'] == 'f'])

employee_procurement = employee_promotions[employee_promotions['department'] == 'Procurement']
procurement_promotion_m = len(employee_procurement[(employee_procurement['gender'] == 'm') & (employee_procurement['is_promoted'])])/len(employee_procurement[employee_procurement['gender']=='m'])
procurement_promotion_f = len(employee_procurement[(employee_procurement['gender'] == 'f') & (employee_procurement['is_promoted'])])/ len(employee_procurement[employee_procurement['gender'] == 'f'])

employee_finance = employee_promotions[employee_promotions['department'] == 'Finance']
finance_promotion_m = len(employee_finance[(employee_finance['gender'] == 'm') & (employee_finance['is_promoted'])])/len(employee_finance[employee_finance['gender'] == 'm'])
finance_promotion_f = len(employee_finance[(employee_finance['gender'] == 'f') & (employee_finance['is_promoted'])])/len(employee_finance[employee_finance['gender'] == 'f'])


employee_hr = employee_promotions[employee_promotions['department'] == 'HR']
hr_promotion_m = len(employee_hr[(employee_hr['gender'] == 'm') & (employee_hr['is_promoted'])]) / len(employee_hr[employee_hr['gender'] == 'm'])
hr_promotion_f = len(employee_hr[(employee_hr['gender'] == 'f') & (employee_hr['is_promoted'])]) / len(employee_hr[employee_hr['gender'] == 'f'])

employee_legal = employee_promotions[employee_promotions['department'] == 'Legal']
legal_promotion_m = len(employee_legal[(employee_legal['gender'] == 'm') & (employee_legal['is_promoted'])]) / len(employee_legal[employee_legal['gender'] == 'm'])
legal_promotion_f = len(employee_legal[(employee_legal['gender'] == 'f') & (employee_legal['is_promoted'])]) / len(employee_legal[employee_legal['gender'] == 'f'])

# -------------------------Graphing------------------------------- 
#Grouping by department  

#Setting Labels for each department
labels = ['Analytics', 'Finance', 'HR', 'Legal', 'Operations', 'Procurement', 'R&D', 'Sales and Management']

#Creating array of all male percentages per department 
men=[analytics_promotion_m, finance_promotion_m, hr_promotion_m, legal_promotion_m, operations_promotion_m,procurement_promotion_m, rd_promotion_m,sale_promotion_m]

#Creating array of all female percentages per department 
women=[analytics_promotion_f, finance_promotion_f, hr_promotion_f, legal_promotion_f, operations_promotion_f,procurement_promotion_f, rd_promotion_f,sale_promotion_f]


X_axis = np.arange(len(labels))
#Graphing for 2 bars per department 
plt.bar(X_axis - 0.2, women, 0.4, label = 'Women', color = '#e8ce95')
plt.bar(X_axis + 0.2, men, 0.4, label = 'Men', color = '#35195e')

#Setting labels for x and y axis
plt.xticks(X_axis, labels)
plt.xlabel("Departments", size = 15) 
#Rotating text for better readability 
plt.xticks(rotation = 45, ha='right')
plt.ylabel("Percent of promotions", size = 15)
plt.title("Percentage of promotions between men and women per department", size = 16)
plt.legend()
plt.show()

```

```python
# Creates seperate tables for male and female employees
employees_male = employee_promotions[employee_promotions.gender == "m"]
employees_female = employee_promotions[employee_promotions.gender == "f"]

# Store number of male/female employees
noMaleEmployees = len(employees_male)
noFemaleEmployees = len(employees_female)

# Filter Sourced into Male and Female
male_sourced = employees_male[employees_male.recruitment_channel == "sourcing"]
female_sourced = employees_female[employees_female.recruitment_channel == "sourcing"]

# Store numbers of
no_male_sourced = len(male_sourced)
no_female_sourced = len(female_sourced)
```

```python
# Identify all departments
departments = ["Sales & Marketing", "Operations", "Technology", "Analytics", "HR","R&D","Procurement","Finance"]

print(len(employee_promotions[employee_promotions.recruitment_channel == "sourcing"])/len(employee_promotions))

men_sourced_proportions = []
women_sourced_proportions = []

#Iterate through departments
for department in departments:
    
    # Calculate proportion sourced
    dept = employee_promotions[employee_promotions.department == department]
    no_sourced = len(dept[dept.recruitment_channel == "sourcing"])
    proportion_sourced_dept = 100 * no_sourced / len(dept)
    
    # Filter for all sourced for a particular department
    no_male_sourced_dept = len(male_sourced[male_sourced.department == department])
    no_female_sourced_dept = len(female_sourced[female_sourced.department == department])
    no_sourced_dept = len(employee_promotions[employee_promotions.recruitment_channel == "sourcing"][employee_promotions.department == department])
    
    # Filter for ratios in departments
    no_male_dept = len(employees_male[employees_male.department == department])
    no_female_dept = len(employees_female[employees_female.department == department])
    dept_total = no_male_dept + no_female_dept
    
    # Calculate proportions per department
    male_proportion_dept = 100 * no_male_dept/dept_total
    female_proportion_dept = 100* no_female_dept/dept_total
    
    #Calculate proportions for sourced department
    male_proportion = 100 * no_male_sourced_dept / no_sourced_dept
    female_proportion = 100 * no_female_sourced_dept / no_sourced_dept
        
    # Output per department
    #print(department + ":")
    #print("%s of men/women in department")
    #print("Men present: ", male_proportion_dept)
    #print("Women present: ", female_proportion_dept)
    #print("% of people sourced: ", proportion_sourced_dept)
    #print("%s of men/women sourced")
    #print("Men Sourced:", male_proportion)
    #print("Women Sourced:", female_proportion)
    #print()
    
    # Append to lists
    men_sourced_proportions.append(male_proportion)
    women_sourced_proportions.append(female_proportion)
```

```python
labels = departments

x = np.arange(len(labels))
width = 0.4

fig, ax = plt.subplots(figsize=(13,8))
ax.bar(labels, women_sourced_proportions, width, label='Women',color="lightcoral")
ax.bar(labels, men_sourced_proportions, width, label="Men", bottom=women_sourced_proportions,color="cornflowerblue")

ax.set_ylabel("Percentage of Sourced Employees")
ax.set_title("Proportions of Men/Women Sourced by each Department")
ax.set_xticks(x, labels)
ax.legend()

plt.show()
```
